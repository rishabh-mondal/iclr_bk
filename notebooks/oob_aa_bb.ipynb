```python
import os

# Path to the directory with label .txt files
labels_dir = "/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data_sentinel/lucknow_sarth_grid_v2/labels"

# List all .txt files in the directory
txt_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

# Iterate over each .txt file and check if it's blank
for file in txt_files:
    file_path = os.path.join(labels_dir, file)
    
    # Check if the file is blank (empty)
    if os.path.getsize(file_path) == 0:
        print(f"Removing blank file: {file}")
        os.remove(file_path)  # Remove the file if it's blank

print("Blank files removed.")
```

```python
import os

label_obb = "/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data_sentinel/cross_val_sentinel_v2/0/train/labels"
label_aa = "/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data_sentinel/lucknow_sarth_grid_v2/labels_aa"
train_aa = "/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data_sentinel/cross_val_sentinel_v2/0/train/labels_aa/"

# Create the destination directory if it doesn't exist
os.makedirs(train_aa, exist_ok=True)

# Iterate through the OBB labels
for file in os.listdir(label_obb):
    print(f"Processing file: {file}")
    
    # Extract the base name (without the .txt extension)
    name = file.split(".txt")[0]
    
    # Define the corresponding AA label file
    aa_file = os.path.join(label_aa, name + ".txt")
    
    # Check if the AA label file exists, if not, skip it
    if not os.path.exists(aa_file):
        print(f"File not found, skipping: {aa_file}")
        continue
    
    # Read the contents of the AA label file
    with open(aa_file, "r") as f:
        lines = f.readlines()
    
    # Write the contents to the destination directory
    with open(os.path.join(train_aa, name + ".txt"), "w") as f:
        for line in lines:
            f.write(line)

print("Processing complete.")
```
