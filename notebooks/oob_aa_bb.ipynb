```python
obb_dir="/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data/amedabad_10_km_buffer_data/labels"
aa_dir="/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data/amedabad_10_km_buffer_data/labels_aa"   
load_dir = obb_dir
save_dir = aa_dir
write_aa_labels_from_obb(load_dir, save_dir)
print(f"Labels written to {save_dir}")
```

```python
label_obb="/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data/multi_state_data_v1/test/labels"
label_aa="/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data/multi_state_data_v1/labels_aa"
train_aa="/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/processed_data/multi_state_data_v1/test_aa/"
os.makedirs(train_aa, exist_ok=True)    
#fetch the labels from the aa labels match name with obb labels and save the aa labels in the train_aa folder
for file in os.listdir(label_obb):
    print(file)
    name = file.split(".txt")[0]
    print(name)
    aa_file = os.path.join(label_aa, name + ".txt")
    print(aa_file)
    with open(aa_file, "r") as f:
        lines = f.readlines()
    with open(os.path.join(train_aa, name + ".txt"), "w") as f:
        for line in lines:
            f.write(line)
```
